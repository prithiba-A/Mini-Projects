```python
import csv

input_file = "/content/Mini Project - Marks Adding.csv"

# Read the input CSV file and calculate the sum of marks for each roll number
roll_marks = {}
with open(input_file, "r") as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        roll_number = row[0]
        marks = int(row[1])
        if roll_number in roll_marks:
            roll_marks[roll_number] += marks
        else:
            roll_marks[roll_number] = marks

# Display each roll number with their total marks
for roll_number, total_marks in roll_marks.items():
    print("Roll Number:", roll_number, "- Total Marks:", total_marks)
```
